```python
import pandas as pd
```

```python
# read the proposals_for_review.xlsx from the data folder
df = pd.read_excel('data/proposals_for_review.xlsx',sheet_name='review')
# filter out the colomns with Ryan Validated? yes or Yes
df = df[df['Flagged for Pedram'].str.lower() == 'yes']
# filter out the columns that Joshua Validated? is not Skipped
df = df[df['Joshua Validated?'].str.lower() != 'skipped']
df.head()
```

```python
# for the theme values that are empty, replace them with theme_as_reviewed value in the same row
df['theme'] = df['theme'].fillna(df['theme_as_reviewed'])
# if theme_as_reviewed is drop, replace with None
df.loc[df['theme_as_reviewed'] == 'drop','theme'] = None
# replace the standard name column that are empty with the latest non-empty value in the same column
df['standard_name'] = df['standard_name'].ffill()
df['sectionid'] = df['sectionid'].ffill()
df['subsectionid'] = df['subsectionid'].ffill()
```

```python
# get list of themes from the column theme
all_themes = df['theme'].unique()
all_themes
```

```python
# let's filter out based on the specific meeting name
metting_name = "qtt_reg_rgc_min__2024-04-03__01"
metting_name = "pwr_mcp_rgc_min__2024-03-21__01"

df0 = df[df['standard_name'] == metting_name+'.pdf']
df0
```

```python
# if voting_result_as_reviewed was not na, then replace the vote_result with the voting_result_as_reviewed with .loc
df0.loc[df0['voting_result_as_reviewed'].notna(),'vote_result'] = df0['voting_result_as_reviewed']

# same for future_date_as_reviewed
df0.loc[df0['future_date_as_reviewed'].notna(),'policy_prop_future_date'] = df0['future_date_as_reviewed']


# group by sectionid and subsectionid and get the policy_prop as a dict
policy_dict = df0.groupby(['sectionid','subsectionid'])['policy_prop'].apply(list).to_dict()
#if the list has one element, then get the first element
for k,v in policy_dict.items():
    policy_dict[k] = [x for x in v if pd.notna(x)]
    if len(policy_dict[k]) == 0:
        policy_dict[k] = float('nan')
        print('WARNING: no policy found for ',k)
    elif len(policy_dict[k]) == 1:
        policy_dict[k] = policy_dict[k][0]
    else:
        for i in range(1,len(policy_dict[k])):
            if policy_dict[k][i] != policy_dict[k][i-1]:
                print('WARNING: different policy found for ',k)
        policy_dict[k] = policy_dict[k][0]
                
policy_dict
```

```python
# group by sectionid and subsectionid and get the theme as a dict only if they are not empty
theme_dict = df0.groupby(['sectionid','subsectionid'])['theme'].apply(list).to_dict()
# remove None values from the theme_dict
for k,v in theme_dict.items():
    theme_dict[k] = [x for x in v if pd.notna(x)]
    if len(theme_dict[k]) == 0:
        theme_dict[k] = float('nan')

theme_dict
```

```python
# group by sectionid and subsectionid and get the vote_result
vote_result_dict = df0.groupby(['sectionid','subsectionid'])['vote_result'].apply(list).to_dict()
# remove nan values from the vote_result_dict
for k,v in vote_result_dict.items():
    vote_result_dict[k] = [x for x in v if pd.notna(x)]
    if len(vote_result_dict[k]) == 0:
        vote_result_dict[k] = None
    else:
        vote_result_dict[k] = vote_result_dict[k][0]
# remove the keys with empty comments
vote_result_dict = {k: v for k, v in vote_result_dict.items() if v is not None}
vote_result_dict
```

```python
import math
# group by sectionid and subsectionid and get the non-empty policy_prop_future_date 
future_date_dict = df0.groupby(['sectionid','subsectionid'])['policy_prop_future_date'].apply(list).to_dict()
# remove all None values from the future_date_dict value list
for k, v in future_date_dict.items():
    future_date_dict[k] = [x for x in v if not pd.isnull(x)]
    if len(future_date_dict[k]) == 0:
        future_date_dict[k] = None
    else:
        future_date_dict[k] = future_date_dict[k][0]


# remove the keys with empty comments
future_date_dict = {k: v for k, v in future_date_dict.items() if v is not None}
future_date_dict
```

```python
# merge the policy_dict and theme_dict, future_date_dict, vote_result_dict into a df with sectionid and subsectionid as index
df1 = pd.DataFrame([policy_dict,theme_dict,future_date_dict,vote_result_dict]).T
df1.columns = ['policy_prop','theme','future_date','vote_result']
df1
```

```python
# read the output file from the data folder
df_output = pd.read_csv('data/output.csv')
df_output = df_output[df_output['meeting_name'] == metting_name]

df_target = pd.read_csv('data/target.csv')
df_target = df_target[df_target['meeting_name'] == metting_name]
```

```python
# find the similar rows by comparing the policy_prop from df1 with full prposals from df_output
# create a list of policy_prop from df1 and output
target_policy_list = df_target['policy_prop'].tolist() 
output_policy_list = df_output['full proposal'].tolist()
print("length of target: {}, length of output: {}".format(len(target_policy_list),len(output_policy_list)))
```

```python
from fuzzywuzzy import fuzz
import numpy as np

def find_top_matching_pairs(list1, list2, threshold=70):
    # Initialize the similarity matrix
    similarity_matrix = np.zeros((len(list2), len(list1)))

    # Fill the similarity matrix with fuzz ratios
    for i, str1 in enumerate(list1):
        for j, str2 in enumerate(list2):
            similarity_matrix[j, i] = fuzz.ratio(str1, str2)

    matches = []
    
    while True:
        # Find the maximum value in the similarity matrix
        max_value = np.max(similarity_matrix)
        
        # If the maximum value is below the threshold, break the loop
        if max_value < threshold:
            break
        
        # Get the indices of the maximum value
        max_index = np.unravel_index(np.argmax(similarity_matrix, axis=None), similarity_matrix.shape)
        
        # Append the indices and the similarity ratio to the matches list
        matches.append((max_index[1], max_index[0], max_value))
        
        # Set the row and column of the maximum value to zero to ignore them in the next iteration
        similarity_matrix[max_index[0], :] = 0
        similarity_matrix[:, max_index[1]] = 0

    return matches
pairs = find_top_matching_pairs(target_policy_list, output_policy_list, 5)
pairs
```

```python
# compare the themes of the target and output 
import re
target_theme_list = df_target['theme'].tolist() 
output_theme_list = df_output['theme'].tolist()

# create confusion matrix by averaging for all the pairs
all_tlist = []
all_olist = []
confusion_matrix_avg = np.zeros((2,2))
for i,j,_ in pairs:
    tlist = target_theme_list[i]
    olist = output_theme_list[j]
    # extract the elements from the string
    tlist = re.findall(r"'(.*?)'", tlist)
    olist = re.findall(r"'(.*?)'", olist)
    all_tlist.extend(tlist)
    all_olist.extend(olist)
    confusion_matrix = np.zeros((2,2))
    # count of true positive: similar themes
    confusion_matrix[0][0] = len(set(tlist).intersection(olist))/max(len(tlist),1)
    # count the false positives: the themes in the target that are not in the output
    confusion_matrix[0][1] = len(set(tlist).difference(olist))/max(len(tlist),1)
    # count the false negatives: the themes in the output that are not in the target
    confusion_matrix[1][0] = len(set(olist).difference(tlist))/max(len(olist),1)
    # count of false negatives: the themes that are not in the target and output
    confusion_matrix_avg += confusion_matrix
confusion_matrix_avg /= len(pairs)
confusion_matrix_avg[1][1] = len(set(all_themes).difference(set(all_tlist).union(set(all_olist))))/len(all_themes)

# make the confusion matrix normalize (out of 100)
confusion_matrix_avg
```
